```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definição do Array das Etapas

atapas = [
    "Revisão bibliográfica e definição do tema",
    "Desenvolvimento do projeto de pesquisa",
    "Coleta e pré-processamento dos dados",
    "Análise exploratória dos dados",
    "Desenvolvimento do modelo de machine learning",
    "Modelos híbridos e Otimização",
    "Explicabilidade e interpretação do modelo",
    "Treinamento do modelo",
    "Validação e Avaliação do modelo",
    "Implementação do modelo em produção",
]


# Definição do Array dos Valores
duracao = [
    (1, 2),  # Revisão bibliográfica e definição do tema
    (3, 4),  # Desenvolvimento do projeto de pesquisa
    (5, 6),  # Coleta e pré-processamento dos dados
    (7, 8),  # Análise exploratória dos dados
    (9, 10),  # Desenvolvimento do modelo de machine learning
    (11, 12), # Modelos híbridos e Otimização
    (13, 14),# Explicabilidade e interpretação do modelo
    (15, 16),# Treinamento do modelo
    (17, 18),# Validação e Avaliação do modelo
    (19, 20) # Implementação do modelo em produção
]

# Criar figura
fig, ax = plt.subplots(figsize=(12, 6))
y_pos = np.arange(len(atapas))

# Plotar barras horizontais
for i, (inicio, fim) in enumerate(duracao):
    ax.barh(y_pos[i], fim - inicio + 1, left=inicio, height=0.6, color='skyblue', edgecolor='black')

# Configurar o eixo y
ax.set_yticks(y_pos)
ax.set_yticklabels(atapas, fontsize=10)
ax.invert_yaxis()  # Inverter o eixo x para que o início do projeto fique à esquerda

# Configurar o eixo x
ax.set_xticks(np.arange(1,22))
ax.set_xticklabels([str(m) for m in range(1,22)])
ax.set_xlabel('Meses', fontsize=12)
ax.set_title('Cronograma do projeto de pesquisa', fontsize=14)


# Exibir a grade
ax.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
